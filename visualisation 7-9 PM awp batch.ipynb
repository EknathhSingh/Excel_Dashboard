```python
#scaling  :  to converts all inputs in same unit 
#use Scaling methods 
#There are 2 types of scaling methods  :
#1. Min Max scaler is also known as normalised 
#2. Standard Scaler  is also known as standardised 
```

```python
#1. MinMaxScaler : -
#formula of MinMaxScaler = (X-X.min())/(X.max()-X.min())
```

```python
#suppose we have 2 input column X1 and X2
X1=np.array([10,20,30,40,50]) #1D array
X2=np.array([100,200,300,400,500])#1D array 
print(X1)
print(X2)
```

```python
def minmaxscaler(X):
    result=(X-X.min())/(X.max()-X.min())
    return result
```

```python
#call function in main program
X1=minmaxscaler(X1)
X2=minmaxscaler(X2)
print("X1: ",X1)
print("X2: ",X2)
```

```python
#2. Standard Scaler  : converts all input columns in same unit
#formula : z-score =(X-mean of(X))/standard deviation(X)
```

```python
#for example : 
#suppose we have 2 input column X1 and X2
X1=np.array([10,20,30,40,50]) #1D array
X2=np.array([100,200,300,400,500])#1D array 
```

```python
#create a user defined function for standard scaler 
def standardscaler(X): 
    result=(X-X.mean())/X.std()
    #std() inbuilt function for standard deviation
    return result
```

```python
#call function in main program
X1=standardscaler(X1)
X2=standardscaler(X2)
print("X1=",X1)
print("X2=",X2)
```
