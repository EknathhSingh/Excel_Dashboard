```python
#seprating input and output from given dataframe
X=df.drop("Price",axis=1) #input select
Y=df['Price'] #output target variable always 1D
```

```python
# Apply Scaling (StandardScaler)
from sklearn.preprocessing import StandardScaler
```

```python
#Create object of StandardScaler class
ss=StandardScaler()
```

```python
#use inbuilt method fit_transform() of StandardScaler class
#Apply
X=ss.fit_transform(X)
```

```python
type(X),X.ndim
```

```python
X
```

```python
#Perform the model means train the model with LinearRegression why?
#bcoz output vriable is a continous numeric
from sklearn.linear_model import LinearRegression
```

```python
#create the object of  LinearRegression class
lr= LinearRegression()
#lr user defined object of  LinearRegression classs
```

```python
#train the model with  LinearRegression class use inbuilt method fit() of  LinearRegression class
lr.fit(X,Y)
```

```python
# to check slope / gradient of predicted line/ best fit line
#use inbuilt attributes coef_ of  LinearRegression class
m=lr.coef_  #m is a user definrd whch hold slope/ gradeient of predicted line
print("slope :",m)
```

```python
# to check intercent of predicted line , use inbuilt attributes intercept_ of   LinearRegression class
c=lr.intercept_
print("Intercept : ",c)
```

```python
# Y=mX+C equation of line
ar=int(input("Enter Area of house in sq ft."))
print("Area : ",ar)
ar=ss.transform([[ar]])
#print("After Scaling Area : ",ar)
#print(type(ar),ar.ndim)

#find the price of house means find the output/target
#Y=mX+c
Y_pred=m*ar+c # X means ar
print("Predicted House price : ",Y_pred)
```
